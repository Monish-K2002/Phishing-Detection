```python
def tokenizer(url):
    tokens = re.split("[/-]",url)
    for i in tokens:
        if i.find("https://")>=0:
            dot_split = i.remove("https://")
            if "https://" in dot_split:
                dot_split.remove("https://")
            if "www" in dot_split:
                dot_split.remove("www")
            
            
            tokens += dot_split
    return tokens
print("\n### Tokenizer defined ###\n")
```

```python
print(test_url)

tokenized_url = tokenizer(test_url)
print(tokenized_url)
```
